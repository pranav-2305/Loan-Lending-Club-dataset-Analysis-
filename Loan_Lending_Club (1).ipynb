# Data cleaning and Transformation

```python
###delete columns with a threshold more than 50 percent
```

```python
df.dropna(thresh=len(df)*0.5, axis=1)
```

```python
df.head()
```

```python
#creates a median at the place of a null at gender
```

```python
df.loc[0,"Gender"]=np.nan 
```

```python

df['Gender'].fillna(df['ApplicantIncome'].median())
```

```python
df['Gender'].fillna(df['ApplicantIncome'].mode()[0])
```

```python
df['Loan_Amount_Term'].fillna(df['LoanAmount'].mean())
```

```python
df.loc[4,"ApplicantIncome"]=np.nan
df.head()
```

```python
df['ApplicantIncome'] = df['ApplicantIncome'].interpolate(method='linear')
```

```python
df.head()
```

```python
df.loc[3,"Loan_Amount_Term"]=np.nan
```

```python
df.head()
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].ffill()

```

```python
df.head()
```

```python
df['Income_MinMax'] = (df['ApplicantIncome'] - df['ApplicantIncome'].min()) / (df['ApplicantIncome'].max() - df['ApplicantIncome'].min())
```

```python
df .head()
```

```python
df['Income_Zscore'] = (df['ApplicantIncome'] - df['ApplicantIncome'].mean()) / df['ApplicantIncome'].std()
```

```python
df.head()
```

```python
df['Income_Log'] = np.log1p(df['ApplicantIncome'])
```

```python
df.head()
```

```python
grouped = df.groupby(['Gender', 'Loan_Status','Education']).agg({
    'ApplicantIncome': ['mean', 'max'],
    'LoanAmount': ['mean', 'sum']
})
```

```python
df['Dependents'] = df['Dependents'].fillna(0)

def change(x):
    if x == '3+':
        return 4
    else:
        return int(x)+1

df['Dependents'] = df['Dependents'].apply(change)

```

# Time Series

```python
df_ = pd.read_excel('loan_level_time_series_dataset.xlsx')
```

```python
df_.head()
```

```python
df_.columns
```

```python
quarterly_avg = df_.groupby(pd.Grouper(key='DisbursalDate', freq='Q'))['LoanAmount'].mean()
quarterly_avg
```

```python
monthly_sum = df_.groupby(pd.Grouper(key='DisbursalDate', freq='M'))['LoanAmount'].sum()
print(monthly_sum)

```

```python


quarterly_avg.plot(kind='line', title='Quarterly Avg Loan Amount')
plt.ylabel("Avg LoanAmount")
plt.xlabel("Quarter")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3-month rolling average of loan amount
df_['LoanAmount_RollingAvg'] = df_['LoanAmount'].rolling(window=3).mean()
```

```python
df_.head()
```

```python
# Exponential Moving Average (EMA) to emphasize recent trends
df_['LoanAmount_EMA'] = df_['LoanAmount'].ewm(span=3, adjust=False).mean()
df_.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(df_.index, df_['LoanAmount'], label='Original')
plt.plot(df_.index, df_['LoanAmount_RollingAvg'], label='3-Month Avg')
plt.plot(df_.index, df_['LoanAmount_EMA'], label='EMA')
plt.title('Loan Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Loan Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Bar plot for original loan amounts
plt.bar(df_.index, df_['LoanAmount'], color='skyblue', label='Original LoanAmount')

# Line plot for EMA
plt.plot(df_.index, df_['LoanAmount_EMA'], color='red', linewidth=2, label='EMA (Smoothed)')

plt.title('Loan Amount with EMA Trend')
plt.xlabel('Date')
plt.ylabel('Loan Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
